```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd
```

# Test conversion pour un fichier

```python
"""
audio_fpath = "./20200413/"
audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
"""

#audio_fpath = "./20200413/"
audio_clips = "breathing-deep.wav"
print("No. of .wav files in audio folder = ",len(audio_clips))

audio_clips

```

```python
x, sr = librosa.load(audio_clips, sr=44100)

print(type(x), type(sr))
print(x.shape, sr)

plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)

X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()

plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
plt.title(audio_clips, loc='center')
plt.savefig(audio_clips+".png")
```

# Test conversion pour un sous dossier


```python
audio_fpath = "./0Rlzhiz6bybk77wdLjxwy7yLDhg1/"
audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
```

```python
size = len(audio_clips)
i = 0
while(i < size):
    
    x, sr = librosa.load(audio_fpath+audio_clips[i], sr=44100)
    
    print(type(x), type(sr))
    print(x.shape, sr)

    plt.figure(figsize=(14, 5))
    librosa.display.waveplot(x, sr=sr)

    X = librosa.stft(x)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()

    plt.figure(figsize=(14, 5))
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar()
    plt.savefig(audio_fpath + audio_clips[i]+".png")
    i = i+1
```
